```python
import pandas as pd


def convert_to_datetime(date_str):
    try:
        return pd.to_datetime(date_str + '-01-01', format='%Y-%m-%d', errors='coerce')
    except:
        return pd.NaT

df['EDAD'] = df['MES_CERTIFICACION'].astype(str).apply(convert_to_datetime)

year_to_filter = 20
state_to_filter = 'JULIO'

print(df.columns)

filtered_df = df[(df['EDAD'].dt.year == year_to_filter) & (df['MES_CERTIFICACION'] == state_to_filter)]
print(filtered_df)
```
